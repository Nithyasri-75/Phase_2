```python
import pandas as pd
import re
df = pd.read_csv("C:/Users/nithy/OneDrive/Desktop/Phase 2/Project 2/Scraping_data/scraped_data.csv")
df.head(5)
```

```python
df.dropna(subset=['job_title', 'skills_required', 'budget_usd'], inplace=True)
df.head(5)
```

```python
def extract_numeric_budget(val):
    try:
        clean_val = re.sub(r'[^\d.]', '', str(val))  
        return float(clean_val)
    except:
        return None
df['budget_usd'] = df['budget_usd'].apply(extract_numeric_budget)
df.dropna(subset=['budget_usd'], inplace=True)
df.head()
```

```python
skill_map = {
    'adobe photoshop': 'photoshop',
    'photoshop': 'photoshop',
    'ps': 'photoshop',
    'js': 'javascript',
    'javascript': 'javascript',
    'py': 'python',
    'python': 'python',
}

def clean_skills(skills_str):
    if pd.isna(skills_str):
        return ''
    skills = re.split(r'[;,|]', skills_str.lower())
    skills = [s.strip() for s in skills if s.strip()]
    normalized = [skill_map.get(s, s) for s in skills]
    return ', '.join(sorted(set(normalized)))  
df['skills_required'] = df['skills_required'].apply(clean_skills)
```

```python
columns_to_clean = ['category', 'job_type', 'location_preference', 'duration_estimate']
for col in columns_to_clean:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
output_path = "C:/Users/nithy/OneDrive/Desktop/cleaned_freelance_jobs.csv"
df.to_csv(output_path, index=False)
print(f"Cleaned data saved to:\n{output_path}")
```
